```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from seaborn import pairplot, scatterplot
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('Dataset/penguins.csv')
```

```python
data.head()
```

```python
data = data.dropna()
print(len(data))
```

```python
# removing year column
data.drop(data.columns[[-1]], axis = 1, inplace = True)
```

```python
species = data['species'].unique()
print(species)
```

```python
data = data[data['species'] != 'Gentoo']   # This limits our species to two
```

```python
pairplot(data, hue='species')
```

```python
# Splitting data to independent and dependent variables

x = data[['bill_length_mm', 'bill_depth_mm']]
y = data['species']
```

```python
# Splitting data into train and test datas

x_train, x_test, y_train, y_test = train_test_split( x, y)
```

```python
model = SVC(kernel = 'linear')
model.fit(x_train, y_train)
```

```python
# Visualizing the linear function for our SVM Classifier

w = model.coef_[0]
b = model.intercept_[0]
x_visual = np.linspace(32,57)
y_visual = -(w[0] / w[1]) * x_visual - b / w[1]

scatterplot(data = x_train, x='bill_length_mm', y='bill_depth_mm', hue=y_train)
plt.plot(x_visual, y_visual)
```

```python
# Testing the accuracy of our model

y_pred = model.predict(x_test)

print(accuracy_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```
